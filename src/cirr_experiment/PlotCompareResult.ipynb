```python
import pandas as pd

from src.cirr_experiment.utils import filter_and_plot_comparison
```

```python
result_clip_r10 = pd.read_csv('../cirr_experiment/clip/recall_at10_clip_vit_g14_cirr.csv', index_col=0)
result_clip_r50 = pd.read_csv('../cirr_experiment/clip/recall_at50_clip_vit_g14_cirr.csv', index_col=0)
result_blip_r10 = pd.read_csv('../cirr_experiment/blip/recall_at10_blip_vit_b_base_cirr.csv', index_col=0)
result_blip_r50 = pd.read_csv('../cirr_experiment/blip/recall_at50_blip_vit_b_base_cirr.csv', index_col=0)
```

# <div style="font-family: 'Garamond', serif; font-size: 22px; color: #ffffff; background-color: #34568B; text-align: center; padding: 15px; border-radius: 10px; border: 2px solid #FF6F61; box-shadow: 0 6px 12px rgba(0, 0, 0, 0.3); margin-bottom: 20px;">Step 2: Plot the comparison</div>

## <div style="font-family: 'Lucida Sans Unicode', sans-serif; font-size: 18px; color: #4A235A; background-color: #D7BDE2; text-align: left; padding: 10px; border-left: 5px solid #7D3C98; box-shadow: 0 4px 6px rgba(0, 0, 0, 0.2); margin-bottom: 10px;">For the Recall@10 metric</div>

```python
filter_and_plot_comparison(
    result_clip_r10,
    result_blip_r10,
    "CLIP (Recall@10)",
    "BLIP (Recall@10)",
    filter_scale=0.1,
    font_size=10,
    cmap="YlGnBu",
    vmin=min(result_clip_r10.min().min(), result_blip_r10.min().min()),
    vmax=max(result_clip_r10.max().max(), result_blip_r10.max().max()),
    save_img_name="clip_blip_recall_at10_cirr.png",
)
```
